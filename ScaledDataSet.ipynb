```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r"DatasetAnalysis_csv.csv")
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaled = MinMaxScaler()
```

```python
data.columns
```

```python
data_scaled = scaled.fit_transform(data[['battery_power', 'clock_speed', 'fc', 'int_memory', 'mobile_wt','n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_h', 'sc_w','talk_time']])
```

```python
data_scaled = pd.DataFrame(data_scaled)
```

```python
data_scaled
```

```python
data_scaled[['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen','wifi', 'price_range']] = data[['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen','wifi', 'price_range']]
```

```python
data_scaled
```

```python
data_scaled.rename(columns={0:'battery_power', 1:'clock_speed', 2:'fc', 3:'int_memory', 4:'mobile_wt',5:'n_cores', 6:'pc', 7:'px_height', 8:'px_width', 9:'ram',10: 'sc_h',11:'sc_w',12:'talk_time'},inplace=True)
```
